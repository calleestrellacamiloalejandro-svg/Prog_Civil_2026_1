# Problema de ingeniería (nivel básico)

Un estudiante de ingeniería necesita **calcular el área de un terreno rectangular** para estimar el espacio de cultivo.

La fórmula es:

[
Área = base \times altura
]

---

# 1. Definir el algoritmo (Entradas – Procesos – Salidas)

### Entradas

Datos que el programa debe recibir:

* base del terreno (metros)
* altura del terreno (metros)

### Procesos

Operaciones que realizará el algoritmo:

* multiplicar base × altura

### Salidas

Resultado que mostrará el programa:

* área del terreno en **m²**

---

# 2. Algoritmo (descripción paso a paso)

1. Iniciar programa
2. Pedir al usuario la **base del terreno**
3. Pedir al usuario la **altura del terreno**
4. Calcular el área usando la fórmula
5. Mostrar el área del terreno
6. Finalizar programa

---

# 3. Pseudocódigo

```
Inicio

Leer base
Leer altura

area = base * altura

Mostrar "El área del terreno es:", area

Fin
```

---

# 4. Código en Python

```python
# Programa para calcular el área de un terreno rectangular

# Entrada de datos
base = float(input("Ingrese la base del terreno en metros: "))
altura = float(input("Ingrese la altura del terreno en metros: "))

# Proceso (cálculo)
area = base * altura

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 5. Ejemplo de ejecución

```
Ingrese la base del terreno en metros: 10
Ingrese la altura del terreno en metros: 5

El área del terreno es: 50 m²
```

## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación **de alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Sus pilares principales son:

* **Legibilidad:** El código es limpio y fácil de leer, lo cual es vital cuando trabajas en proyectos colaborativos de ingeniería.
* **Versatilidad:** Tiene bibliotecas potentes para cálculo numérico (`NumPy`), análisis de datos (`Pandas`) y visualización (`Matplotlib`).
* **Interpretado:** No necesitas compilar el código; se ejecuta línea por línea, lo que facilita mucho las pruebas rápidas en **Google Colab**.

## Un Ejemplo con Contexto de Ingeniería

Imagina que necesitas calcular el **volumen de agua** necesario para regar una parcela rectangular. En lugar de usar una calculadora cada vez, podemos crear un pequeño script.

Aquí tienes un ejemplo de cómo se vería la lógica en Python. Fíjate bien en los nombres de las variables:

````python
# Definición de variables con nombres descriptivos
largo_parcela_m = 50.0
ancho_parcela_m = 30.0
lamina_riego_mm = 12.0  # Altura de agua requerida

# Lógica del cálculo
# 1 mm de lámina equivale a 1 litro por metro cuadrado (1 l/m2)
area_m2 = largo_parcela_m * ancho_parcela_m
volumen_total_litros = area_m2 * lamina_riego_mm

print(f"Para un área de {area_m2} m2, necesitas {volumen_total_litros} litros de agua.")
````

### ¿Por qué este código es "bueno"?

1. **Variables descriptivas:** Cualquiera que lea `largo_parcela_m` entiende qué mide y en qué unidades está (metros).
2. **Comentarios:** Explican el "por qué" de las operaciones.
