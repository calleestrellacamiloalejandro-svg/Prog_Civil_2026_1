# **Dosificación de Mezcla para Columnas**

**Planteamiento**:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

1. Solicitar la altura de la columna en metros.

2. Solicitar el diámetro de la columna en metros.

3. Calcular el volumen de la columna: $$v = \pi * r^2 * h$$ (Sugerencia: usar 3.1416 o importar math.pi).

4. Calcular la cantidad de sacos: $$sacos = v * 9.5$$


**Prueba del programa**:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

```python
 # ENTRADA: Solicitar la altura de la columna en metros.
 # Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d / 2
v = 3.1416 * r**2 * h
sacos = v * 9.5

#SALIDA: Imprime el resultado
print("El volumen de la columna es:", round(v, 3), "m3.")
print("Se requieren", round(sacos, 2), "sacos de cemento.")


```

Aquí está la explicación de cada línea del código de cálculo:

*   `r = d / 2`: Esta línea calcula el radio de la columna (`r`) dividiendo el diámetro (`d`) entre 2. Es un paso necesario para calcular el volumen de un círculo.

*   `v = 3.1416 * r**2 * h`: Esta línea calcula el volumen de la columna (`v`). Utiliza la fórmula para el volumen de un cilindro: \( \pi \cdot r^2 \cdot h \). Aquí, `3.1416` es una aproximación de \( \pi \), `r**2` es el radio al cuadrado, y `h` es la altura de la columna.

*   `sacos = v * 9.5`: Esta línea calcula la cantidad total de sacos de cemento necesarios (`sacos`). Se multiplica el volumen de la columna (`v`) por la cantidad de sacos requeridos por metro cúbico (9.5 sacos/m³).

*   `print("El volumen de la columna es:", round(v, 3), "m3.")`: Esta línea imprime el volumen calculado de la columna. Usa `round(v, 3)` para mostrar el volumen redondeado a tres decimales para mayor claridad.

*   `print("Se requieren", round(sacos, 2), "sacos de cemento.")`: Esta línea imprime la cantidad final de sacos de cemento necesarios. Usa `round(sacos, 2)` para mostrar el número de sacos redondeado a dos decimales.
