# **Métodos y cálculos**
+ Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72*(T_{bs}+T_{bh})+40.6$$

Donde:

1. $T_{bs}$ = Temperatura de bulbo seco ($^oC$)

2. $T_{bh}$ = Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$ ambiente no estresante
2. Sí $ITH > 70 \space and \space ITH <= 78$ situación de alerta por estrés térmico
3. Sí $ITH > 78 \space and \space ITH <= 83$ peligro por estrés térmico
4. Sí $ITH > 83$ situación crítica

## **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere a las condiciones ambientales en las que los animales no experimentan estrés por calor ni por frío, lo que les permite mantener su temperatura corporal central dentro de un rango normal con un gasto mínimo de energía. La zona de confort térmico, o zona de termoneutralidad, es el rango de temperaturas ambientales donde el ganado puede mantener su temperatura corporal sin alterar significativamente sus tasas metabólicas.

Cuando las condiciones ambientales se desvían de esta zona, los animales deben activar mecanismos fisiológicos y conductuales para disipar o conservar calor. El estrés por calor es una preocupación particular en regiones tropicales y subtropicales, ya que afecta negativamente la producción de leche, la ganancia de peso, la reproducción y la salud general de los animales. El Índice de Temperatura y Humedad (ITH) es una herramienta ampliamente utilizada para evaluar el nivel de estrés térmico en el ganado, combinando los efectos de la temperatura del aire y la humedad relativa.

```python
#ENTRADA: pedir tbs (Temperatura de bulbo seco) y tbh (Temperatura de bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo: "))

#PROCESO: Calculo de ith (Indice de Temperatura y Humedad)
ith = 0.72 * (tbs + tbh) + 40.6

#SALIDA: Imprimir el resultado, de la evaluación del ith y el confort térmico
if ith <= 70:
  print("El ambiente es no estresante")
elif ith > 70 and ith <= 78:
  print("La situación es de alerta por estrés térmico")
elif ith > 78 and ith <= 83:
  print("La situación es de peligro por estrés térmico")
else:
  print("La situación es crítica")
```
